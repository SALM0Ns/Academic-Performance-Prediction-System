```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv('Grade Predice.csv')
```

```python
df.columns = df.columns.str.strip()
```

```python
grade_dummies = pd.get_dummies(df['Grade Fundamental Programming Concepts'], prefix='Grade')
df = pd.concat([df.drop(columns=['Grade Fundamental Programming Concepts']), grade_dummies], axis=1)
```
